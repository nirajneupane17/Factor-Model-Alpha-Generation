```python
import sys; sys.path.insert(0,"../src")
import pandas as pd, matplotlib.pyplot as plt
from portfolio_optimizer import max_sharpe_weights, risk_parity_weights, performance_metrics
print("OK")
```

```python
f=pd.read_csv("../data/monthly_factors.csv",index_col="Date",parse_dates=True)
factors=["MKT_RF","SMB","HML","RMW","CMA"]
ms=max_sharpe_weights(f,factors); rp=risk_parity_weights(f,factors)
for nm,w in [("Max Sharpe",ms),("Risk Parity",rp),("Equal Weight",{c:0.2 for c in factors})]:
    port=sum(w.get(c,0)*f[c] for c in factors)
    pm=performance_metrics(port)
    print(f"{nm}: Sharpe={pm['sharpe_ratio']}  Ann={pm['ann_return_pct']}%  MaxDD={pm['max_drawdown_pct']}%")
```
